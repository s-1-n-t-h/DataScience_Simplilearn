## Problem Statement
     You have been provided with a dataset that lists Ohio State's leading causes of detah from the year 2012.
     
     Using the two data points cause of death and percentile, draw a pie chart to visulaize the dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cause = ['Chronic Diseases','Unintentional Injuries','Aizheimers',
         'Infuenza and Pneumonia','Sepsis','Others']
```

```python
percentile = [62, 5, 4, 2, 1, 26]
```

```python
# set the pie chart properties
```

```python
# set figure size
plt.figure(figsize=(10,10))

# explode the largest pie in the data set
explode = (0.05,0,0,0,0,0)

# set pie chart properties
plt.pie(percentile,labels=cause,explode=explode,autopct='%1.1f%%',startangle=80)

# set axis equal to draw pie as circle
plt.axis('equal')

# set title of the pie chart
plt.title('Ohio Sate - 2012: Leading Causes of Death')
# show the plot (pie chart)
```
