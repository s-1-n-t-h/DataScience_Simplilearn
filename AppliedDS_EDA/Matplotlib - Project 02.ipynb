```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cause = ['Chronic Disease','Unintentional Injuries','Alzehmeimers','Influenza and Pneumonia','Sepsis','Others']
```

```python
percentile = [62, 5, 4, 2, 1, 26]
```

```python
plt.figure(figsize=(10,10))
explode = (0.05,0,0,0,0,0)
plt.pie(x=percentile,labels=cause,explode=explode,startangle=85)
plt.axis('equal')
plt.title('Ohio State - 2012: Leading Causes of Death')
plt.show()
```
