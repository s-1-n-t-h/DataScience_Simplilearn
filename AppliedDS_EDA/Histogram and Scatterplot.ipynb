```python
style.use('ggplot')
plt.figure(figsize=(7,7))
plt.hist(y,bins=50) # binning to intervals
plt.xlabel('price in 10000s USD')
plt.ylabel('number of houses')
plt.show()
```

```python
# scatter plot

style.use('ggplot')
plt.figure(figsize=(7,7))
plt.scatter(boston_real_state_data.data[:,5],boston_real_state_data.target)
plt.xlabel('price in 10000s USD')
plt.ylabel('number of houses')
plt.show()
```
