```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('../datasets/auto-mpg.csv')
```

```python
dataset.head()
```

```python
# write a user defined function for origin

def origin(num):
    if num==1:
        return 'USA'
    elif num==2:
        return 'Europe'
    else:
        return 'Asia'
    
# use apply function

dataset['origin'] = dataset['origin'].apply(origin)
```

```python
dataset.head()
```

```python
# draw the pair plot using sns for mpg, weight, origin and with hue origin, set the size to 4
# note: hue is variable in data set to map plot aspects to different colors
```

```python
sns.pairplot(dataset[['mpg','weight','origin']],hue='origin',size=4)
```
