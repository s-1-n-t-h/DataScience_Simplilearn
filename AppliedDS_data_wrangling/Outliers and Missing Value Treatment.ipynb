```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(data=np.c_[dataset['data'], dataset['target']],columns=dataset['feature_names']+['target'])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
for column in df:
    plt.figure()
    sns.boxplot(df[column])
```

```python
for column in df:
    plt.figure()
    df.boxplot([column])
```
